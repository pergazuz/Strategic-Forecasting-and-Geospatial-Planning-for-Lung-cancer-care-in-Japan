```python
import pandas as pd
import numpy as np
import os
import geopandas as gpd
from shapely.geometry import Point
import matplotlib
import matplotlib.pyplot as plt
import folium
import plotly_express as px
from datetime import datetime
import geemap
# from ipygee import*
```

```python
jpnShp = gpd.read_file('data/gadm41_JPN_1.shp')
```

```python
jpnShp
```

```python
jpnCsv = pd.read_csv('data/gadm41_JPN_1.csv')
jpnCsv
```

```python
ax = jpnShp.plot(figsize=(14, 14))
jpnShp.apply(lambda x: ax.annotate(text=jpnCsv.NAME_1,xy=x.geometry.centroid.coords[0], ha='center', color = 'black', size = 9),axis=1)
jpnShp.plot(ax = ax, edgecolors='black')
plt.title('Administrative level 1 map in Japan', fontsize=16)
plt.show();
```

```python
# Plot the map
ax = jpnShp.plot(figsize=(14, 14))

# Annotate each region
for idx, row in jpnShp.iterrows():
    # Get the corresponding name from the CSV
    name = jpnCsv.iloc[idx]['NAME_1']
    # Get the centroid of the geometry
    centroid = row.geometry.centroid
    # Annotate the map
    ax.annotate(text=name, xy=(centroid.x, centroid.y), ha='center', color='black', size=6)

# Plot the shapefile with edge colors
jpnShp.plot(ax=ax, edgecolor='black')

# Set the title
plt.title('Administrative level 1 map in Japan', fontsize=16)

# Show the plot
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the CSV file
csv_path = 'data/lung_cancer.csv'
lung_cancer_df = pd.read_csv(csv_path)

merged = jpnShp.merge(lung_cancer_df, left_on='name', right_on='AREA')

# Plot the map with the frequency color of each city
fig, ax = plt.subplots(1, 1, figsize=(14, 14))
merged.plot(column='NUMBER', ax=ax, legend=True, cmap='OrRd', edgecolor='black')

# Annotate each region
for idx, row in merged.iterrows():
    name = row['name']
    centroid = row['geometry'].centroid
    ax.annotate(text=name, xy=(centroid.x, centroid.y), ha='center', color='black', size=6)

# Set the title
plt.title('Lung Cancer Frequency by Region in Japan (2018)', fontsize=16)

# Show the plot
plt.show()
```
